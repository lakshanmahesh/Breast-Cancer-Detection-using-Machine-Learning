```python
# convert ? --> NaN
data['bare_nucleoli'] = data['bare_nucleoli'].replace('?',np.nan)
```

```python
data = data.dropna()
```

```python
# convert Dtype Object --> int
data['bare_nucleoli'] = pd.to_numeric(data['bare_nucleoli'])
```

```python
data.info()
```

```python
data['class'].value_counts()
```

```python
data['class'] = data['class'].replace(2,0)
data['class'] = data['class'].replace(4,1)
```

```python
data['class'].value_counts()
```

#### (0 for benign, 1 for malignant)

```python
x = data.drop(['id','class'],axis=1)
y= data['class']
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaler = scaler.fit_transform(x)
x = pd.DataFrame(data_scaler,columns=x.columns)
```

```python
x.head()
```

```python
y.value_counts()
```

#### imbalance data

```python
import matplotlib.pyplot as plt
y.value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Class Distribution')
plt.xlabel('Classes')
plt.ylabel('Count')
plt.xticks([0, 1], ['benign', 'malignant'], rotation=0)
plt.show()
```

```python
from imblearn.combine import SMOTEENN
combined = SMOTEENN(sampling_strategy='auto')
x,y = combined.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
y.value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Class Distribution')
plt.xlabel('Classes')
plt.ylabel('Count')
plt.xticks([0, 1], ['benign', 'malignant'], rotation=0)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict,StratifiedKFold
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
score= cross_val_score(knn_model,x,y,cv=5).mean()
print(score)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
score=cross_val_score(rf_model,x,y,cv=5).mean()
print(score)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
score=cross_val_score(nb_model,x,y,cv=5).mean()
print(score)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
score=cross_val_score(lr_model,x,y,cv=8).mean()
print(score)
```

### Support Vector Machine

```python
from sklearn.svm import SVC
svc_model = SVC()
score=cross_val_score(svc_model,x,y,cv=8).mean()
print(score)
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier()
score=cross_val_score(gb_model,x,y,cv=8).mean()
print(score)
```

```python
from sklearn.model_selection import GridSearchCV
param_grids = { 'weights': ['uniform', 'distance'],
                'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
                'n_neighbors':[2,5,6,7,8]}

grid_obj = GridSearchCV(estimator=knn_model,param_grid=param_grids)
best_model = grid_obj.fit(x,y)
best_model.best_params_
   
       
```

```python
score=cross_val_score(best_model,x,y,cv=8).mean()
print(score)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = cross_val_predict(best_model, x, y )
cm = confusion_matrix(y, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
import pickle
with open('model.pickle','wb') as file:
    pickle.dump(best_model,file)
```
